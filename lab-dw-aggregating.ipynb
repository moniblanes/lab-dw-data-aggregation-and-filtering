1. Create a new DataFrame that only includes customers who:
   - have a **low total_claim_amount** (e.g., below $1,000),
   - have a response "Yes" to the last marketing campaign.

```python
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
import pandas as pd
df = pd.read_csv(url)
print(df.head())
```

```python
# create a new dataframe with clients whose total claim amount is below 1000 & response = 'Yes'
new_df = df[(df['Total Claim Amount'] < 1000) & (df['Response'] == 'Yes')]
print(new_df.head())
new_df.shape
```

2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
pivot1 = pd.pivot_table(
    new_df,
    values=["Monthly Premium Auto","Customer Lifetime Value","Total Claim Amount"],
    index=['Policy Type','Gender'],
    aggfunc="mean",
    fill_value=0,
).round(2)
pd.options.display.float_format = '{:,.2f}'.format
pivot1


# Insights:
# 1. Customers with 'Personal Auto' policy type tend to have higher monthly premiums compared to 'Corporate Auto' and 'Special Auto'.
# 2. Female customers generally have a higher average Customer Lifetime Value than male customers across both policy types. 
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
df['State'].value_counts()
df['Customer'].value_counts()
pivot2= pd.pivot_table(     
    new_df, 
    values="Customer",
    index="State",
    aggfunc="count"
 ).round(2)
pivot2
pivot2[pivot2['Customer'] > 500]
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
pivot3= pd.pivot_table(     
    new_df, 
    values="Customer Lifetime Value",
    index=["Education", "Gender"],
    aggfunc=["median", "max", "min"]
    ).round(2)
pd.options.display.float_format = '{:,.2f}'.format
pivot3
```

Insights:
 1. Customers with 'Doctorate' education level have the highest median Customer Lifetime Value, indicating that higher education may correlate with higher spending.
 2. Male customers with 'Doctorate' education have the highest maximum Customer Lifetime Value, suggesting that this demographic may include high-value customers.
 3. Customers with 'High School' education level have the lowest minimum Customer Lifetime Value, indicating a potential area for targeted marketing or customer retention strategies.
 4. Overall, there is a noticeable variation in Customer Lifetime Value across different education levels and genders

5. The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
print(df.columns)
#extract the month from the 'Effective To Date' column and create a new column 'Month'
new_df['Effective To Date'] = pd.to_datetime(new_df['Effective To Date'])   
new_df['Month'] = new_df['Effective To Date'].dt.month
print(new_df[['Effective To Date', 'Month']].head())

pivot5 = pd.pivot_table(
    new_df,
    values="Policy",
    index="State",
    columns="Month",
    aggfunc="count"
).round(2)
pivot5
```

6.  Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

*Hint:*
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
#group by 'State' and 'Month' to count the number of policies sold in each state per month and sort in descending order
grouped = new_df.groupby(['State', 'Month']).size().reset_index(name='Policy Count')
grouped_sorted = grouped.sort_values(by='Policy Count', ascending=False)    
print(grouped_sorted)

#new dataframe with only customers from 'California' ,'Oregon' and 'Arizona' containgn the number of policies sold per month for each state
states_df = grouped_sorted[grouped_sorted['State'].isin(['California', 'Oregon', 'Arizona'])]
print(states_df)    
```

7. The marketing team wants to analyze the effect of different marketing channels on the customer response rate.

Hint: You can use melt to unpivot the data and create a table that shows the customer response rate (those who responded "Yes") by marketing channel.

External Resources for Data Filtering: https://towardsdatascience.com/filtering-data-frames-in-pandas-b570b1f834b9

```python
# use melt to unpivot the data and create a table that shows the customer response rate (those who responded "Yes") by sales channel.
melted = pd.melt(
    new_df,
    id_vars=['Sales Channel', 'Response'],
    value_vars=['Customer'],
    var_name='Metric',
    value_name='Count'
)
pivot6 = pd.pivot_table(
    melted[melted['Response'] == 'Yes'],
    values='Count',
    index='Sales Channel',
    aggfunc='count'
).round(2)

pivot6
```
